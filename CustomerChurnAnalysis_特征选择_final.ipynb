## 3.RFE法筛选变量

递归消除特征法使用一个基模型来进行多轮训练，每轮训练后，移除若干权值系数的特征，再基于新的特征集进行下一轮训练。RFE通过递归减少考察的特征集规模来选择特征。首先，预测模型在原始特征上训练，每个特征指定一个权重。之后，那些拥有最小绝对值权重的特征被踢出特征集。如此往复递归，直至剩余的特征数量达到所需的特征数量。

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
model = LogisticRegression() 
rfe = RFE(model,15) 
rfe = rfe.fit(X_train,y_train)
print(rfe.support_) 
print(rfe.ranking_)
```

```python
a = rfe.ranking_.tolist()
for each in zip(FE,a):
    print(each)
```

## 4.Filter法筛选变量

SelectKBest会移除得分前 k 名以外的所有特征(取top k)。这种方法主要基于特征与目标变量之间的相关性，可以通过scores_判断特征的重要性。

```python
from sklearn.feature_selection import SelectKBest, f_classif
selector = SelectKBest(f_classif, k=15) #第一个参数表示检验方法，可以使用chi2, f_classif|第二个参数k表示取出几个特征
a=selector.fit(X_train,y_train)
print(np.array(a.scores_),'\n',a.get_support())
```
